```python
# import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.colors as mcolors
from itertools import groupby

##############################################################################
# 1. YOUR DATA
##############################################################################
data = {
    'Pathways:': 7767,
    'Trophic type:': 1473,
    'Chemical:': 275939 + 59 - 5672 - 9729 - 910,
    'Enzymatic Activity:': 85324,
    'Cell shape:': 33437,
    'Cell dimensions:': 3074+3099,
    'Motility:': 7930,
    'Sporulation:': 9591,
    'Pigment:': 99,
    'Gram stain:': 14726,
    'GC content:': 7963,
    'pH conditions:': 16006,
    'Temp. conditions:': 26555,
    'Temperature preference:': 30244,
    'NaCl preference:': 304,
    'NaCl conditions': 13473,
    'Oxygen preference:': 42745,
    'Pathogen:': 2639,
    'Antibiotic:': 5672 + 9729,
    'Assay': 119094,
    'Isolation source:': 174261 + 17288 + 2802 + 78 + 910,
    'UniProt - Gene Ontology - Microbes': 332114889,
    'UniProt - Reactions - Microbes': 22814571 ,
    'UniProt - Enzymatic Activity - Microbes': 29038183 , 
    'UniProt - Chemical - Microbes': 18140307,
    'UniProt - Gene Ontology - Human': 884215,
    'UniProt - Reactions - Human': 28304,
    'UniProt - Enzymatic Activity - Human': 19231, 
    'UniProt - Chemical - Human': 11503,
        
}

##############################################################################
# 2. MAP KEYS TO THEMES (same as before)
##############################################################################
def get_theme(trait):
    #trait = trait.rstrip(':')  # remove trailing colon
    print(trait)
    
    metabolism = [
        'Pathways',
        'Trophic type',
        'Chemical',
        'Enzymatic Activity',
        'Assay',
        'Antibiotic'
    ]
    
    morphology = [
        'Cell shape',
        'Cell dimensions',
        'Sporulation',
        'Pigment',
        'Gram stain',
        'GC content',
        'Motility'
    ]
    
    environmental = [
        'pH conditions',
        'Temp. conditions',
        'Temperature preference',
        'NaCl preference',
        'NaCl conditions',
        'Oxygen preference'
    ]
    
    ecology = [
        'Pathogen',
        'Isolation source'
    ]
    
    uniprot_microbes = [
        'UniProt - Gene Ontology - Microbes',
        'UniProt - Reactions - Microbes',
        'UniProt - Enzymatic Activity - Microbes',
        'UniProt - Chemical - Microbes'
    ]
    
    uniprot_human = [
        'UniProt - Gene Ontology - Human',
        'UniProt - Reactions - Human',
        'UniProt - Enzymatic Activity - Human',
        'UniProt - Chemical - Human'
    ]

    
    if trait in environmental:
        return 'Environmental'
    elif trait in metabolism:
        return 'Metabolic'
    elif trait in morphology:
        return 'Morphological'
    elif trait in ecology:
        return 'Ecological'
    elif trait in uniprot_microbes:
        return 'UniProt - Microbes'
    elif trait in uniprot_human:
        return 'UniProt - Human'
    else:
        return 'Other'  # fallback

# Build a list of (theme, trait, count)
data_list = []
for key, count in data.items():
    trait = key.rstrip(':')
    theme = get_theme(trait)
    data_list.append((theme, trait, count))

print(data_list)
# Group by theme, then sort each group by descending count
data_list.sort(key=lambda x: x[0])  # groupby needs grouped data
sorted_data_list = []
for theme, group in groupby(data_list, key=lambda x: x[0]):
    group_list = list(group)
    group_list.sort(key=lambda x: x[2], reverse=True)
    sorted_data_list.extend(group_list)

data_list = sorted_data_list

##############################################################################
# 3. DEFINE THEME ORDER (left->right)
##############################################################################
desired_theme_order = [
    'Environmental',
    'Metabolic',
    'Morphological',
    'Ecological',
    'UniProt - Microbes',
    'UniProt - Human'
]

def theme_order_sort(item):
    theme = item[0]
    if theme not in desired_theme_order:
        return 999
    else:
        return desired_theme_order.index(theme)

data_list.sort(key=theme_order_sort)

themes  = [t for (t, _, _) in data_list]
traits  = [tr for (_, tr, _) in data_list]
counts  = [c for (_, _, c) in data_list]

print(themes)

##############################################################################
# 4. CREATE x-POSITIONS AND COLORS
##############################################################################
theme_positions = {theme: i for i, theme in enumerate(desired_theme_order)}

print(theme_positions)
x_positions = [theme_positions[t] for t in themes]

xscale = 0.8
x_positions_scaled = [x * xscale for x in x_positions]

# We'll define a color map so each theme has a distinct color.
colors = plt.cm.tab10.colors[:6]
theme_color_map = {
    'Environmental': colors[0],
    'Metabolic':     colors[1],
    'Morphological': colors[2],
    'Ecological':    colors[3],
    'UniProt - Microbes':    colors[4],
    'UniProt - Human':    colors[5],
    
    # 'Other':         'gray'
}
point_colors = [theme_color_map.get(t, 'gray') for t in themes]

# y-positions (top to bottom)
y_positions = np.arange(len(traits))

##############################################################################
# 5. GLOBAL LOG10 SCALING FOR BUBBLES
##############################################################################
#  - We find the global min & max count
#  - We'll map from log10(min_count)->log10(max_count) onto
#    a chosen marker-size range, e.g. [20, 300].

min_count = min(counts)
max_count = max(counts)

log_min = math.log10(min_count)
log_max = math.log10(max_count)

marker_size_min = 20  # size (pts^2) for the smallest bubble
marker_size_max = 400 # size (pts^2) for the largest bubble

sizes = []
for c in counts:
    log_c = math.log10(c)
    # Normalize to [0,1]
    norm = (log_c - log_min) / (log_max - log_min) if log_max != log_min else 0
    # Scale to [marker_size_min, marker_size_max]
    size_val = marker_size_min + norm * (marker_size_max - marker_size_min)
    sizes.append(size_val)

##############################################################################
# 6. MAKE THE PLOT
##############################################################################
plt.figure(figsize=(5, 7))

plt.scatter(
    x_positions_scaled,
    y_positions,
    s=sizes,        # globally scaled sizes
    alpha=0.8,
    c=point_colors,
    clip_on=False
)

# Hide x-axis ticks/labels
plt.gca().set_xticks([])
plt.gca().set_xticklabels([])

# Label y-axis with traits
plt.yticks(ticks=y_positions, labels=traits, fontsize=8)
plt.tick_params(axis='y', which='major', pad=1)

# Adjust vertical space
extra_padding = 0.6
plt.ylim(y_positions[0] - extra_padding, y_positions[-1] + extra_padding)

# Horizontal padding
left_extra  = 0.2
right_extra = 1.4
plt.xlim(min(x_positions_scaled) - left_extra, max(x_positions_scaled) + right_extra)

##############################################################################
# 7. OFFSET BUBBLE LABELS
##############################################################################
# We'll offset the text based on the bubble radius so it doesn't overlap.
# Note that marker areas (s=...) are in points^2, so:
# radius in points = sqrt(area / pi).
# We'll scale that to data coords by an additional factor.
##############################################################################
offset_factor = 3  # tweak to suit

for (x, y, c, s_) in zip(x_positions_scaled, y_positions, counts, sizes):
    # s_ is the marker area in pt^2; approximate radius:
    r_pts = math.sqrt(s_ / math.pi)
    # Convert to "data" x-units approximately:
    # We'll assume ~100 points ~ 1 data unit, i.e. 1 inch at default DPI.
    # This is approximate but usually good enough. Adjust multiplier as you like.
    r_data_approx = r_pts / 100.0  

    # Then multiply by an offset factor so the text is beyond the edge
    x_offset = r_data_approx * offset_factor
    plt.text(
        x + x_offset,
        y,
        str(c),
        va='center',
        ha='left',
        fontsize=8,
        clip_on=False
    )

##############################################################################
# 8. BUBBLE REFERENCE LEGEND
##############################################################################
# We'll show reference markers for the min_count and max_count (and maybe some middle values).
# Let's pick 4 reference counts: min_count, ~some middle, ~some bigger, and max_count.
##############################################################################

reference_counts = [min_count, 1000, 10000, max_count]
reference_sizes = []
for rc in reference_counts:
    log_rc = math.log10(rc)
    norm_rc = (log_rc - log_min) / (log_max - log_min) if log_max != log_min else 0
    size_rc = marker_size_min + norm_rc * (marker_size_max - marker_size_min)
    reference_sizes.append(size_rc)

legend_handles = [
    plt.scatter([], [], s=s, color='gray', alpha=0.8, clip_on=False)
    for s in reference_sizes
]
legend_labels = [f"{c}" for c in reference_counts]

legend1 = plt.legend(
    handles=legend_handles,
    labels=legend_labels,
    title="Counts (log scale)",
    labelspacing=2,
    loc="upper left",
    fontsize=8,
    title_fontsize=9
)
plt.gca().add_artist(legend1)

##############################################################################
# 9. THEMES LEGEND
##############################################################################
theme_handles = []
for theme in desired_theme_order:
    color = theme_color_map.get(theme, 'gray')
    handle = mlines.Line2D([], [], marker='o', linestyle='',
                           color=color, markersize=6, label=theme)
    theme_handles.append(handle)

plt.legend(
    handles=theme_handles,
    title='Themes',
    loc='lower right',
    fontsize=8,
    title_fontsize=9,
    labelspacing=0.5
)

plt.tight_layout(pad=1)
plt.savefig('bubble_plot_v2.pdf', format='pdf')
plt.savefig('bubble_plot_v2.png', format='png', dpi=300)
plt.show()
```
